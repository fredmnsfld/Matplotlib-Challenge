```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#Store results in DataFrame
data1_df = pd.DataFrame(mouse_metadata)
data2_df = pd.DataFrame(study_results)
```

```python
# Checking the number of mice.
data1_df.nunique()
```

```python
data2_df.nunique()
```

```python
# Combine the data into a single dataset
combined_data_df = pd.merge(data1_df,data2_df,
                                 how='outer', on='Mouse ID')
```

```python
# Display the data table for preview
combined_data_df.head()
```

```python
#Check count
combined_data_df.count()
```

```python
#Check for Nan and null values
combined_data_df.isnull().sum().sum()
```

```python
#Create datafram based on subset for all rows where timepoint = 45 so we can compare results of each drug regiment by the final timepoint which is 45.
comb45 = combined_data_df[combined_data_df.Timepoint <45] 
comb45
```

## Summary Statistics

```python
# Generate a summary statistics table of mediamean, n, var, stand deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
comb45_mean_df = comb45.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
comb45_mean_df = comb45_mean_df.rename("mean")
comb45_mean_df
```

```python
comb45_var_df = comb45.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
comb45_var_df = comb45_var_df.rename("variance")
comb45_var_df
```

```python
comb45_std_df = comb45.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
comb45_std_df = comb45_std_df.rename("variance")
comb45_std_df
```

```python
comb45_sem_df = comb45.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
comb45_sem_df = comb45_sem_df.rename("variance")
comb45_sem_df
```

```python
comb_45_summary = pd.concat([comb45_mean_df, comb45_var_df, comb45_std_df,comb45_sem_df], axis=1, join='inner')
comb_45_summary
```
